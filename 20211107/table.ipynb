```python
import numpy as np
import jax.numpy as jnp
```

```python
highDetEarning = jnp.array(np.load("constant/highIncomeDetermined.npy"))
lowDetEarning = jnp.array(np.load("constant/lowIncomeDetermined.npy"))
```

```python
# deterministic income
detEarning = jnp.array(np.load("constant/detEarningHigh.npy"))
############################################################################################################ low skill feature 
# fix the deterministic income
Alow = jnp.concatenate([detEarning[:46]*0.5, detEarning[46:]-45])
Ahigh = jnp.concatenate([detEarning[:46], detEarning[46:]-25])
```

```python
plt.plot(Ahigh, label = "high artificial")
plt.plot(Alow, label = "low artificial")
plt.plot(highDetEarning, label = "high real")
plt.plot(lowDetEarning, label = "low real")
plt.legend()
```
