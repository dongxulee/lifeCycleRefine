```python
%pylab inline
import numpy as np
import pandas as pd
matplotlib.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size': 20})
```

```python
ages = list(range(21, 50))
housingDecisionBoundary_noRetirement = np.load("H_boundary_full_model.npy")
housingDecisionBoundary = np.load("H_boundary_without_401k_model.npy")
plt.title("Housing purchase behavior: wealth level decision boundary")
plt.plot(ages,housingDecisionBoundary_noRetirement,'*-', label = "Model without 401k feature")
plt.plot(ages, housingDecisionBoundary,'*-',  label = "Model with 401k feature")
plt.xlabel("age level")
plt.ylabel("wealth level")
plt.legend()
plt.savefig("housingDecisionBoundary")
```

```python
investmentRatio = np.load("investmentRatio.npy")
savingRatio = np.load("savingRatio.npy")
investmentRatio_without_housing = np.load("investmentRatio_without_housing.npy")
savingRatio_without_housing = np.load("savingRatio_without_housing.npy")
investmentRatio_without_401k = np.load("investmentRatio_without_retirement.npy")
savingRatio_without_401k = np.load("savingRatio_without_retirement.npy")
```

```python
plt.title("owner_to_be agent's decision rules on investment ratio across models")
plt.plot(ages, investmentRatio, label = "Model with both housing and 401k")
plt.plot(ages, investmentRatio_without_401k, label = "Model without 401k")
plt.plot(ages, investmentRatio_without_housing, label = "Model without housing")
plt.xlabel("age level")
plt.ylabel("ratio")
plt.legend()
```

```python
plt.title("owner_to_be agent's decision rules on investment ratio across models")
plt.plot(ages, savingRatio, label = "Model with both housing and 401k")
plt.plot(ages, savingRatio_without_401k, label = "Model without 401k")
plt.plot(ages, savingRatio_without_housing, label = "Model without housing")
plt.xlabel("age level")
plt.ylabel("ratio")
plt.legend()
```
