```python
%pylab inline
import numpy as np
import pandas as pd
matplotlib.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size': 10})
```

```python
ages = list(range(21, 50))
housingDecisionBoundary_noRetirement = np.load("H_boundary_full_model.npy")
housingDecisionBoundary = np.load("H_boundary_without_401k_model.npy")
plt.title("Housing purchase behavior: wealth level decision boundary")
plt.plot(ages,housingDecisionBoundary_noRetirement,'*-', label = "Model without 401k feature")
plt.plot(ages, housingDecisionBoundary,'*-',  label = "Model with 401k feature")
plt.xlabel("age level")
plt.ylabel("wealth level")
plt.legend()
plt.savefig("housingDecisionBoundary")
```

```python
wealthGap = [5,10,15,20,25,30,35,40,50]
for wealthgap in wealthGap:
    plt.figure()
    plt.title("Investment Ratio" + "_" + str(wealthgap))
    plt.plot(ages,np.load("investmentRatio_" +str(wealthgap) + ".npy"), label = "withHousing")
    plt.plot(ages,np.load("investmentRatio_without_housing" +str(wealthgap) + ".npy"), label = "withoutHousing")
    plt.legend()
```

```python
wealthGap = [5,10,15,20,25,30,35,40,50]
for wealthgap in wealthGap:
    plt.figure()    
    plt.title("Saving Ratio_" + str(wealthgap))
    plt.plot(ages, np.load("savingRatio_" +str(wealthgap) + ".npy"), label = "withHousing")
    plt.plot(ages, np.load("savingRatio_without_housing" +str(wealthgap) + ".npy"), label = "withoutHousing")
    plt.legend()
```
